```python
import pandas as pd
```

```python
df = pd.read_csv('Econ_Charactristics_Hamilton_Tracts.csv')
df
```

```python
tracts = []
for i in range(len(df)):
    if(df['Label (Grouping)'][i] != "\xa0\xa0\xa0\xa0Estimate" and df['Label (Grouping)'][i] != "\xa0\xa0\xa0\xa0Percent"):
        x = df['Label (Grouping)'][i].split(',')[0]
        x = x.split('-> ')[1]
        tracts.append(x)

tracts
```

```python
import geopandas as gpd
```

```python
df2 = gpd.read_file('./Tennessee Census Tracts/tl_2016_47_tract.shp')
tracts2 = df2['NAMELSAD'].tolist()
```

```python
final = set(tracts2).intersection(tracts)
```

```python
values = pd.DataFrame()
percent = pd.DataFrame()
```

```python
c = 0
for i in range(len(df)):
    if(df['Label (Grouping)'][i] != "\xa0\xa0\xa0\xa0Estimate" and df['Label (Grouping)'][i] != "\xa0\xa0\xa0\xa0Percent"):
        x = df['Label (Grouping)'][i].split(',')[0]
        x = x.split('-> ')[1]
        if(x in final):
            values = values.append(df.loc[i+1], ignore_index=True)
            values['Label (Grouping)'][c] = x
            percent = percent.append(df.loc[i+2], ignore_index=True)
            percent['Label (Grouping)'][c] = x
            c+=1
```

```python
values.rename(columns={'Label (Grouping)': 'NAMELSAD'}, inplace=True)
percent.rename(columns={'Label (Grouping)': 'NAMELSAD'}, inplace=True)
```

```python
values
```

```python
percent
```

```python
res_val = pd.merge(values, df2, on='NAMELSAD')
res_val
```

```python
res_perc = pd.merge(percent, df2, on='NAMELSAD')
res_perc
```

```python
res_val['TRACTCE'] = res_val['TRACTCE'].astype(int)
res_perc['TRACTCE'] = res_perc['TRACTCE'].astype(int)
```

```python
df2 = pd.read_csv("tract_transit_score_home.csv")
df2.rename(columns={'TRACTCE_home': 'TRACTCE'},inplace=True)
df2
```

```python
final_val = pd.merge(res_val, df2, on='TRACTCE')
final_val
```

```python
final_perc = pd.merge(res_perc, df2, on='TRACTCE')
final_perc
```

```python
final_val['COUNTYFP'] = final_val['COUNTYFP'].astype(int)
final_perc['COUNTYFP'] = final_perc['COUNTYFP'].astype(int)
```

```python
final_val = final_val.loc[final_val['COUNTYFP']==65]
final_perc = final_perc.loc[final_perc['COUNTYFP'] == 65]
```

```python
final_val = final_val.reset_index(drop = True)
final_perc = final_val.reset_index(drop = True)
```

```python
final_val.to_csv("Hamilton_Estimates.csv")
final_perc.to_csv("Hamilton_Percents.csv")
```
